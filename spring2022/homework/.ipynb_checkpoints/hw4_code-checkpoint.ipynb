```python
import numpy as np
import math

def dp_mechanism(data, b, epsilon):
    n = len(data)
    data = data.T
    Sxy = np.dot(data[0], data[1])
    Sxx = np.dot(data[0], data[0])
    num_noise = np.random.laplace(loc=0, scale=4*b*b/epsilon, size=None)
    denom_noise = np.random.laplace(loc=0, scale=2*b*b/epsilon, size=None)
    num = clamp(Sxy + num_noise, (-n*b*b, n*b*b))
    denom = clamp(Sxx + denom_noise, (b/1000, n*b*b))
    return num/denom

def calculate_ols(data):
    data = data.T
    Sxy = np.dot(data[0], data[1])
    Sxx = np.dot(data[0], data[0])
    return Sxy / Sxx
```

```python
def clamp(x, bounds):
    """Replace any x_i less than lower with lower, 
           and any x_i greater than upper with upper."""
    return np.clip(x, *bounds)

def generate_data(n):
    x_s = np.random.uniform(-0.5,0.5,n)
    noise = np.random.normal(0, np.sqrt(0.02), size=n)
    y_s = x_s + noise
    y_s = clamp(y_s, (-1, 1))
    res = np.vstack((x_s, y_s)).T
    return res
```

```python
def bias(betas):
    return np.mean(betas-1)
    
def std_dev(betas):
    return np.std(betas)
```

```python
epsilon = 0.1
b = 1
num_trials = 10000

def run_trial(n):
    beta_hats = np.zeros(num_trials)
    beta_squiggles = np.zeros(num_trials)
    for i in range(num_trials):
        data = generate_data(n)
        beta_hats[i] = calculate_ols(data)
        beta_squiggles[i] = dp_mechanism(data, b, epsilon)

    return bias(beta_hats), bias(beta_squiggles), std_dev(beta_hats), std_dev(beta_squiggles)
    
```

```python
import matplotlib.pyplot as plt

bias_hats = np.zeros(50)
bias_squiggles = np.zeros(50)
std_dev_hats = np.zeros(50)
std_dev_squiggles = np.zeros(50)

i = 0
for n in range(100, 5001, 100):
    a, b, c, d = run_trial(n)
    bias_hats[i] = a
    bias_squiggles[i] = b
    std_dev_hats[i] = c
    std_dev_squiggles[i] = d
    i += 1

# hats is ols
# squiggles is dp

plt.plot(range(100, 5001, 100), bias_hats, color='deepskyblue', label="ols_bias")
plt.plot(range(100, 5001, 100), std_dev_hats, color='lightpink', label="ols_std_dev")

plt.title("OLS Bias and Standard Deviation")
plt.xlabel("Value of n")
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('hw4_plot_ols.png')
plt.show()

plt.plot(range(100, 5001, 100), bias_squiggles, color='deepskyblue', label="dp_bias")
plt.plot(range(100, 5001, 100), std_dev_squiggles, color='lightpink', label="dp_std_dev")

plt.title("DP Bias and Standard Deviation")
plt.xlabel("Value of n")
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('hw4_plot_dp.png')
plt.show()
```

```python
plt.plot(range(100, 5001, 100), clamp(bias_hats, (-1, 1)), color='deepskyblue', label="ols_bias")
plt.plot(range(100, 5001, 100), clamp(std_dev_hats, (-1, 1)), color='lightpink', label="ols_std_dev")

plt.title("OLS Bias and Standard Deviation")
plt.xlabel("Value of n")
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('hw4_plot_ols_clamped.png')
plt.show()

plt.plot(range(100, 5001, 100), clamp(bias_squiggles, (-1, 1)), color='deepskyblue', label="dp_bias")
plt.plot(range(100, 5001, 100), clamp(std_dev_squiggles, (-1, 1)), color='lightpink', label="dp_std_dev")

plt.title("DP Bias and Standard Deviation")
plt.xlabel("Value of n")
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('hw4_plot_dp_clamped.png')
plt.show()
```
